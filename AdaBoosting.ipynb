```python
import pandas as pd
```

```python
dataset = pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset = pd.get_dummies(dataset,dtype=int)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent = dataset[['age', 'bmi', 'children', 'sex_female', 'sex_male','smoker_no', 'smoker_yes']]
independent
```

```python
dependent = dataset[['charges']]
dependent
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.ensemble import AdaBoostRegressor
aboostregressor= AdaBoostRegressor(estimator=None, n_estimators=500, learning_rate=1.0, loss='exponential', random_state=None)
aboostregressor.fit(x_train,y_train)
```

```python
aboostregressor.feature_names_in_
```

```python
aboostregressor.estimator_errors_
```

```python
y_predict=aboostregressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
rscore= r2_score(y_test,y_predict)
rscore
```
